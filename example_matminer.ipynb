```python
import numpy as np
import pandas as pd

from matminer.featurizers.conversions import StrToComposition
from matminer.featurizers.composition import ElementProperty
from matminer.featurizers.structure import DensityFeatures
from pymatgen.core.structure import Structure
import pymatgen

import warnings
warnings.simplefilter("ignore")
```

```python
! pip show pymatgen
```

```python
data = {'composition_formula': ['SrTiO3']}
df = pd.DataFrame(data)
df
```

```python
df = StrToComposition().featurize_dataframe(df, col_id="composition_formula", ignore_errors=True)
df
```

```python
df = ElementProperty.from_preset(preset_name="magpie").featurize_dataframe(df, col_id="composition", ignore_errors=True)
df
```

```python
with open('/content/SrTiO3.cif', 'r') as f:
    cif_content = f.read()
```

```python
crystal_tmp = Structure.from_str(cif_content, fmt = "cif")
crystal_tmp
```

```python
df['cif'] = [crystal_tmp]
df
```

```python
df = DensityFeatures().featurize_dataframe(df, col_id="cif", ignore_errors=True)
df
```
